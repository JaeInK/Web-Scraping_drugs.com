```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.drugs.com')
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
top_100 = soup.find('ul', attrs={'class' : 'top100list column-span column-span-3'})
```

```python
top_100_list = []
for tag in top_100.find_all('a'):
    top_100_list.append(tag.text)
top_100_list
```

```python
len(top_100_list)
```

```python
def checkGName1(drug_name):
    r = requests.get('https://www.drugs.com/'+drug_name+'.html')
    soup = BeautifulSoup(r.text, 'html.parser')
    temp = soup.find('p', attrs={'class' : 'drug-subtitle'})
    if(temp.find('a') == None):
        g_name = temp.contents[1].strip()
    else:
        g_name = temp.find('a').text.strip()
    print(drug_name,g_name)
```

```python
for name in top_100_list:
    checkGName1(name)
```

```python
def checkGName3(drug_name):
    r = requests.get('https://www.drugs.com/'+drug_name+'.html')
    soup = BeautifulSoup(r.text, 'html.parser')
    temp = soup.find('p', attrs={'class' : 'drug-subtitle'})
    if(' ' in temp.contents):
        temp.contents.remove(' ')
    if(temp.contents[1].name != 'a'):
        g_name = temp.contents[1].strip().split(" ")[0].strip()
    else:
        g_name = temp.contents[1].text.strip()
        
    g_check = g_name.split(' ')
    if(len(g_check) != 1):
        g_name = g_check[0] + '-' + g_check[2]
        
    if(g_name.lower() == drug_name.lower()):
        url = 'https://www.drugs.com/comments/'+drug_name+'.html/'
    else:
        url = 'https://www.drugs.com/comments/'+g_name+'/'+drug_name+'.html/'
    print(url)
```

```python
for name in top_100_list:
    checkGName3(name)
```

```python
def Scraping(drug_name, records):
    r = requests.get('https://www.drugs.com/'+drug_name+'.html')
    soup = BeautifulSoup(r.text, 'html.parser')
    temp = soup.find('p', attrs={'class' : 'drug-subtitle'})
    if(' ' in temp.contents):
        temp.contents.remove(' ')
    if(temp.contents[1].name != 'a'):
        g_name = temp.contents[1].strip().split(" ")[0].strip()
    else:
        g_name = temp.contents[1].text.strip()
        
    g_check = g_name.split(' ')
    if(len(g_check) != 1):
        g_name = g_check[0] + '-' + g_check[2]
        
    if(g_name.lower() == drug_name.lower()):
        url = 'https://www.drugs.com/comments/'+drug_name+'.html/'
    else:
        url = 'https://www.drugs.com/comments/'+g_name+'/'+drug_name+'.html/'
    
    for page_num in range(1, 500):
        r = requests.get(url+'?page='+str(page_num))
        soup = BeautifulSoup(r.text, 'html.parser')
        user_comments = soup.find_all('div', attrs={'class' : 'user-comment'})
        if(len(user_comments)==0):
            break

        for i in range(len(user_comments)):
            user_comment = user_comments[i]
            pain = user_comment.contents[0].find('b')
            if(pain!=None):
                pain = pain.text.strip()
            comment = user_comment.contents[0].find('span').text.strip()
            user = user_comment.find('p', attrs={'class' : 'user-name'}).contents[0].strip()
            rating = user_comment.find('div', attrs={'class' : 'rating-score'})
            if(rating!=None):
                rating = rating.text.strip()
            records.append((drug_name,pain,comment,user,rating))
            if(i==500):
                print("Error: Too many interation. Duplicate data inserted.")
```

```python
for index, row in df.iterrows():
    if(len(row['user'])<user_length):
        how_many_u = user_length - len(row['user'])
        user_modified = row['user'] + how_many_u*'0'
        df.set_value(index, 'user', user_modified)
    if(len(row['drug'])<user_length):
        how_many_d = user_length - len(row['drug'])
        drug_modified = row['drug'] + how_many_d*'0'
        df.set_value(index, 'drug', drug_modified)
        
df
```
